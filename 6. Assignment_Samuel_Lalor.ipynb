7-1. Add a new column to df7, named 'LA'.<br> 
Assign a value of 1 to this column for players who are a member of either the LA Angels or LA Dodgers teams.<br>
For all other players, set the value in the 'LA' column to 0.<br>
Note that for players in the LA teams, their value on the Tm column is LAA or LAD.

7-2. Answer whether the batting average (BA) for players on the LA teams (Angels and Dodgers) is higher compared to that of the players from all other teams.<br>

Note that applying mean() function to the BA column and comparing the outcome values will give you a wrong answer.<br> That will give you the "average" of "Batting Average" (it is an average of average) rather than the actual batting average for each team.<br>

Instead, to accurately compute the BA for each team, first get the total of 'H' (Hits) and 'AB' (At Bats) for each group.<br>
Then, compute BA by dividing the total Hits by the total At Bats for each team.<br>
Finally, compare these calculated means to draw a conclusion."

```python
# Answer
la_players = df7[df7['Tm'].isin(['LAA', 'LAD'])]

total_hits_la = la_players['H'].sum()
total_at_bats_la = la_players['AB'].sum()

total_hits_other = df7.loc[~df7['Tm'].isin(['LAA', 'LAD']), 'H'].sum()
total_at_bats_other = df7.loc[~df7['Tm'].isin(['LAA', 'LAD']), 'AB'].sum()

ba_la = total_hits_la / total_at_bats_la
ba_other = total_hits_other / total_at_bats_other

if ba_la > ba_other:
    conclusion = "The LA teams (Angels and Dodgers) have a higher batting average."
else:
    conclusion = "The LA teams (Angels and Dodgers) do not have a higher batting average."

conclusion
```
