3. Write a python program that checks the following conditions and prints texts:<br>
    If green is entered to the input function – print Car is allowed to go<br>
    If yellow, – print Car has to wait<br>
    If red, – print Car has to stop<br>
    If it is unrecognized signal, print Unrecognized signal

```python
signal = input("What is a traffic signal? Enter either red, yellow, or green: ")
```

```python
# Answer

signal = signal.lower()

if signal == 'g':
    print('Car is allowed to go')
elif signal == 'y':
    print('Car has to wait')
elif signal == 'r':
    print('Car has to stop')
else:
    print('Unrecognized signal')

```

5. A company sells their product for $10.<br>
There are quantity discounts (see the following table).<br>

| Quantity    | Discounts |
| -------- | ------- |
| 5 - 10  | 10%    |
| 11 - 20 | 20%     |
| 21 -     | 30%    |

    Write a program that displays the total amount of the purchase after the discount based on the variable, quantity.

```python
# Assume that users enter only a number.
quantity = input('How many product do you want to purchase: ')
quantity = int(quantity)
```

```python
# Answer

if quantity >= 5 and quantity <= 10:
      total = quantity * 10 * 0.9 
elif quantity >= 11 and quantity <= 20:
      total = quantity * 10 * 0.8
elif quantity >= 21:
      total = quantity * 10 * 0.7

print(f'Your total is ${total}')

```

• If the person is at least 13 years old, but less than 20 years old, he or she is a teenager.<br>
